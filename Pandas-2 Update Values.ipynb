# Update & modifying row & columns & data value

```python
import pandas
```

```python

d3={'a':[1,2,3],'b':[11,22,33],'c':[111,222,333]}
df=pandas.DataFrame(d3)
df
```

```python
#to check the coloumn index
col_x=df.columns
#to_check the row index
row_x=df.index
print(col_x,row_x)
```

```python
#Rename columns index
df.columns=['aa','bb','cc']
df
```

```python
#Rename row_index
df.index=[11,22,33]
df
```

```python
#Rename all columns index name as upper case: by using list comprehension
df.columns=[x.upper() for x in df.columns]
df.columns
```

```python
#Rename all columns index name:key as old value and value as new value of coumns index
df.rename(columns={'AA':'a','BB':'b','CC':'c'},inplace=True)
df
```

```python
#Update and modify any data
#first we access that data which we want to update, end assign value to it by pressing equal to
#loc and iloc method is best for accessing the data
```

```python
df.loc[1,'a']=10
df

df.loc[[1,2],'a']=20
df

df.loc[[1,2],'a']=[30,40]
df

df.loc[1,:]=2
df
```

```python
#If you want to change data according to particular value, it can be possible by filter
filt=(df['a']==2)
df.loc[filt,'c']=500
df
```

## Apply, Map, applymap, replace Methods

## apply

#### it is a method which allow to apply other methods or own functions/lambda function on every elements of series object type , if you apply this(apply method)  on dataframe type then apply method will work on series level


```python
df=pandas.read_csv(r"C:\Users\DELL\Desktop\kPython RND folder\Python pandas\survey_results_public.csv")
ndf=df.head(10)
d3={'a':[1,2,3],'b':[11,22,33],'c':[111,222,333]}
df3=pandas.DataFrame(d3)
ndf
```

```python
#Ex-1 How to get numbers of charactors( len property) of particular series object
ndf['Country'].apply(len)
```

```python
#Ex-2 How i can apply own function on series or dataframe object
def get_lenght_value(Country):
    return Country.upper()

ndf['Country'].apply(get_lenght_value)

#note:-
```

```python
def get_lenght_value1(Country,l):
    return Country.upper(), str(len(l))

x,y=get_lenght_value1('abhilash','kkk')
```

```python
#we can also use lamda function in apply method
ndf.loc[:,'Country'].apply(lambda x:x.lower())
```

```python
#we can update country coloumn by doing this
ndf.loc[:,'Country']=ndf.loc[:,'Country'].apply(lambda x:x.lower())
ndf
```

```python
#we can use custom function in lambda function by doing this
ndf["Country"].apply(lambda x:get_lenght_value1(x,x))
```

```python
#If use apply method on dataframe object type then it will work on series level,rather than work on every every element 
ndf.apply(len)
#here respondent have 10 rows, so givs 10
#if want to get length of columns of every row(series) of dataframe object then you will need to pass axis parameter in along with apply method
ndf.apply(len,axis='columns')
```

## map

### it used for subsitute the element in series data object
it have one specility that if you don't provided new element in series data type object then it will replace with 'nan'

```python
df3.loc[:,'a'].map({1:'abh',2:'chi'})
```

## replace
it used for subsitute the element in series data object
it have a specility that if you don't provided new element in series data type object then old element will be there

```python
df3.loc[:,'b'].replace({11:'abh',22:'chi'})
```

# Adding Columns, Rows and Single Elements to DataFrame 

```python
dict_src={'First Name':['shivam','chitresh','abhilash'],'Last Name':['rana','pratap','singh'],'ID':['sss','ccc','aaa']}
df=pandas.DataFrame(dict_src)
df
```

```python
# How to add new column in DataFrame with combination of first name and last name column:
df['Full Name']=df['First Name']+' '+df['Last Name']
df
```

```python
# How to delete column and row in DataFrame: By drop method with passing parameter columns(for columns) and index(for rows)
df.drop(columns=['First Name','Last Name'],inplace=True)
```

```python
# How to Split existing columns and resulted value assign with new columns
df[['First Name','Last Name']]=df['Fulll Name'].str.split(' ',expand=True)
```

```python
# How to add single data to DataFrame: using append method
df.append({'First Name':'Gungun'},ignore_index=True)
df
```

```python
df2=pandas.DataFrame({'First Name':['Gungun','ravi'],'Last Name':['rana','pratap'],'ID':['ggg','rrr']})
df2
```

```python
#How to add two dataframe into single dataframe : this will be do by append method with using ignor index equal to true
# in append method there is no parameter of like inplace, so in order to permanent change in DataFrame we have to assign this change into new variable d3
df3=df.append(df2,ignore_index=True)

```

```python
# How to delete columns and rows with using filter(means condition)
filt=(df3['First Name']=='ravi')
df3.drop(index=df3[filt].index,inplace=True)
df3
```

# Sorting Data

```python
dict_src={'First Name':['shivam','chitresh','abhilash'],'Last Name':['rana','pratap','singh'],'ID':['sss','ccc','aaa']}
df=pandas.DataFrame(dict_src)
df
```

```python
# How to sort data by First Name
df.sort_values(by='Last Name')
```

```python
# How to sort multiple sorting
df.sort_values(by=['First Name','Last Name'],ascending=False)
```

```python
df.sort_values(by=['First Name','Last Name'],ascending=[True,True],inplace=True)
```

```python
# How to sort data by index
df.sort_index()
```

```python
# How to sort data in series object
df['First Name'].sort_values()
```
